```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

from matplotlib import style
from matplotlib.pyplot import figure
style.use('ggplot')
```

```python
v = np.arange(start = 0, stop = 4, step = 0.1)
n = len(v)

capacity = np.zeros(n);

for i in range(n):
    capacity[i] = (1000 * v[i]) -2000;

figure(1, figsize=(16, 6)) 
plt.plot(v, capacity)
plt.xlabel("Napięcie [V]")
plt.ylabel("Pojemność [%]")
#plt.savefig("./png/koh-h2o.png")
#plt.savefig("./pdf/koh-h2o.pdf")
plt.show()
```
