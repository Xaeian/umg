```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
freq = [100, 1000, 10000, 100000, 200000] # Hz
z_mod = [0.030, 0.032, 0.313, 3.023, 5.98] # ohm
fi = [-56.3, 64.6, 85.90, 88.67, 88.86] # rad
ls = [-39.6, 4.66, 4.97, 4.81, 4.764] # uH
esr = [0.0166, 0.0138, 0.0701, 0.1194, 0.12] # ohm
```

```python
plt.subplots_adjust(hspace=0.4)
t = np.arange(0.01, 20.0, 0.01)

plt.subplot(221)
plt.semilogx(freq, z_mod)
plt.title('z_mod')
plt.grid(True)

plt.subplot(222)
plt.semilogx(freq, fi)
plt.title('fi')
plt.grid(True)

plt.subplot(223)
plt.semilogx(freq, ls)
plt.title('ls')
plt.grid(True)

plt.subplot(224)
plt.semilogx(freq, esr)
plt.title('esr')
plt.grid(True)

plt.savefig('eclz-mostek.pdf')
plt.show()
```
