```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

from matplotlib import style
from matplotlib.pyplot import figure
style.use('ggplot')
```

```python
T = [0, 10, 20, 25, 40, 60, 80, 100, 120, 140]
H2O = [97.6, 102.4, 112.4, 117.9, 135.3, 147.5, 162.5, 179.3, 206, 367]

H20_22C5 = (H2O[2] + H2O[3]) / 2
np.disp(H20_22C5)

figure(1, figsize=(16, 6)) 
plt.plot(T, H2O)
plt.xlabel("Temperatura [°C]")
plt.ylabel("Rozpuszczalność KOH [g]")
plt.savefig("./png/koh-h2o.png")
plt.savefig("./pdf/koh-h2o.pdf")
plt.show()
```

```python
mKOH = H20_22C5 # masa KOH [g]
mH20 = 100 # masa wody [g]
pc = (100 * mKOH) / (mKOH + 100) # stężenie procentowe

uK = 39 # masa molowa potasu [g]
uO = 16 # masa molowa tlenu [g]
uH = 1 # masa molowa modoru [g]
uOH = uO + uH # masa molowa wody [g]
uKOH = uK + uO + uH

# uKOH --> mKOH
# uK   --> mK
# uOH  --> mOH

mK = uK * mKOH / uKOH
mOH = uOH * mKOH / uKOH
molOH100ml = mOH / uOH # ilość moli OH- w 1kg wody

molOH = 10 * molOH100ml
pOH = -math.log10(molOH)
pH = 14 - pOH

np.disp(pH)
```

```python
mKOH = np.arange(start = 0.1, stop = H20_22C5, step = 0.1)
n = len(mKOH)

pc = np.zeros(n)
for i in range(n):
    pc[i] = (100 * mKOH[i]) / (mKOH[i] + 100)

molOH = np.zeros(n)
pOH = np.zeros(n)
pH = np.zeros(n)

for i in range(n):
    molOH[i] = 10 * mKOH[i] / uKOH
    pOH[i] = -math.log10(molOH[i])
    pH[i] = 14 - pOH[i]

figure(1, figsize=(16, 6))
plt.plot(pc, pH)
plt.xlabel("Temperatura [°C]")
plt.ylabel("Rozpuszczalność KOH [g]")
plt.savefig("./png/koh-ph.png")
plt.savefig("./pdf/koh-ph.pdf")
plt.show()
```
